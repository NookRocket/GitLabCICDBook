# Section 2 : Basic CI/CD Workflows with GitLab

Continuous Integration(CI)
* Practice of continuously integrating code changes
* Ensure that the project can still be built / compiled
* Ensure that any changes pass all tests, guidelines, and code compliance standards

   **Advantages**
   - Error are deteced early in the development process (debug as soon as possible)
   - Reduce integration problems
   - Allow developers to work faster because it can be easily detected problem

Continuous Delivery(CD)
* Ensure that the software can be deployed anytime to production
* Commonly, the latest version is deployed to a testing or staging system (pre-production)

    **Advantages of Continuous Delivery and Continuous Deployment**
    - Ensure that every change is releasable by testing that it can be deployed
    - Reduced risk of a new deployment
    - Delivers value much faster

### Managing Secrets

* DO NOT keep any secrets, credentials, usernames in pipline or project files
* Define environment variables that will be available when running the pipeline to store the secrets

* Uncheck `Protect variable` to allow all branchs in the repository can use it
* `Masked and hidden` to keep in high level secret like password or token
